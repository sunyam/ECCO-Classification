## Some data analysis before the annotation process:

```python
import os
import pandas as pd
import numpy as np
```

```python
# Checking if we have same category names:
path = '../txtLAB/Doc Class 5.0/ECCO_TrainingData_Trial4/'

print(len(os.listdir(path)[2:]))

categories = ['Agriculture.csv', 'Botany.csv', 'Church.csv', 'Commerce.csv', 
              'Drama.csv', 'Fiction.csv', 'History.csv', 'History_Natural.csv', 'Law.csv', 'Mathematics.csv', 
              'Medicine.csv', 'Physics.csv', 'Poetry.csv', 'Politics.csv', 'Rhetoric.csv', 'Sermons.csv', 
              'Travels.csv']
print(set(categories) == set(os.listdir(path)[2:]))
```

#### Creating a table for all 17 classes:

```python
df_s = []
for cat in os.listdir(path)[2:]:
    df = pd.read_csv(path+cat)
    df['Class Label'] = cat[:-4]
    print(cat)
    print(df.shape)
    df_s.append(df)
```

```python
table1_all_17 = pd.concat(df_s, ignore_index=True)
print(table1_all_17.nunique())
table1_all_17.head()
```

#### Creating a table for ALL the data we have:

```python
path = '../txtLAB/Koustuv_ECCO_Metadata/'

df_s2 = []
for cat in os.listdir(path):
    df = pd.read_csv(path+cat, delimiter='\t')
    df['TableName'] = cat[:-4]
    print(cat)
    print(df.shape)
    df_s2.append(df)
```

```python
table2_all_metadata = pd.concat(df_s2, ignore_index=True)
print(table2_all_metadata.nunique())
table2_all_metadata.head()
```
